```python
# Downloaded the data as an Excel file from Analytics dashboard: https://datastudio.google.com/u/0/reporting/1jGAALdF8Mx3e3RxvkHyEUE5Iju0tY2j2/page/RXJQB
# Saved all the data as text (copied it into notepad and then pasted it back in using Text Import Wizard)
# Renamed the sheetname as "Searched Terms" and the file location is "C:\Users\satal\Desktop\Internship\UK Trade Tariffs_Page Views and Searches Table.xlsx"

"""
Import Excel
"""

import pandas as pd 

ExcelRead = pd.read_excel(r"C:\Users\satal\Desktop\Internship\UK Trade Tariffs_Page Views and Searches Table.xlsx", sheet_name = "Searched Terms")

print(ExcelRead)
```

```python
#retrieved JSON file from data.jason from uktrade/global-uk-tariff (github)
#uploaded JSON file to python using json.load

import json

with open('/Users/satal/Desktop/Internship/data.json') as data_file:
        data = json.load(data_file)
        
for code in data:
    code["total"] = 0

print(data[0])       
```

```python
# matches everything in the searched terms list to all the possibile commodity codes it could match to
# you can double check this code works by using the UKGT tool: https://www.check-future-uk-trade-tariffs.service.gov.uk/tariff
# the code also counts the matches. For example, if 17028937 is the comm code for 'vegtables', and 17028937 has been searched for 50 times while 'vegetables' has been searched for 20 times. Then the total search for 17028937 is 70. 

for row in ExcelRead.itertuples():
    index, term, count = row
    #if index>500:
      #  break
   # print (term)
    for code in data:
        match = term in code["commodity"] or term.lower() in code["description"].lower() or term.lower() in code["change"].lower()
        if match :
            code["total"] += count
            

            
```

```python
#max(data, key = lambda code : code["total"])

#sorts and gives the top 10 most searched terms

top_searches=sorted(data, key = lambda code : code["total"])[-10:]

#print(top_searches)

#saved the top 10 searches as a JSON file so that we can start representing the data as diagrams using Excel

with open ('top_searches.json','w') as f:
    json.dump(top_searches, f, indent=2)
```

```python
top_100=sorted(data, key = lambda code : code["total"])[-100:]

with open ('top_100.json','w') as p:
    json.dump(top_100, p, indent=2)
```
